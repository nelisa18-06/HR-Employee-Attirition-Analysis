# Project Title:
HR Employee Attrition Analysis

Objective:
The objective of this project is to analyze employee attrition and understand the key factors that influence why employees leave an organization. Using HR data, the project explores how variables such as department, job role, overtime, experience, and income are related to employee attrition.

This project was done to practice and demonstrate end-to-end data analysis using Python, SQL, and Power BI.


# Dataset Description

The dataset contains employee-level HR information such as:

Demographic details (Age, Gender)

Job-related attributes (Department, JobRole, JobInvolvement)

Work conditions (OverTime, WorkLifeBalance)

Compensation details (MonthlyIncome)

Experience metrics (YearsAtCompany, TotalWorkingYears)

Target variable: Attrition (Yes / No)


Each row in the dataset represents one employee.

# Tools Used

Python (Jupyter Notebook) – Data cleaning, preprocessing, and initial analysis

Pandas – Handling and manipulating data

SQL (SQLite) – Querying and aggregating data for analysis

Power BI – Creating visualizations and dashboard for insights

# Key Insights

Certain departments and job roles experience higher employee attrition

Employees working overtime show higher attrition counts

Employees with fewer years of experience tend to leave more frequently

Lower average monthly income is associated with higher attrition
